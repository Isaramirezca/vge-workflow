## Open Drone Map

Open Drone Map (ODM) is a free and open-source software package that is used to process and generate 3D models, point clouds, and orthomosaic maps from drone or UAV imagery. It is a comprehensive photogrammetry tool that can automatically generate detailed 3D models and maps from aerial photographs captured by drones or other unmanned aircraft. ODM is built on top of several open-source libraries and tools, including OpenCV, OpenSfM, and the Python programming language. ODM can be used for a variety of applications, including land surveying, agriculture, forestry, environmental monitoring, and construction. It can help to reduce the cost and time required for mapping and surveying projects, while also providing accurate and detailed results. Recently ODM released user-friendly application and API named WebODM, this application provides a web interface to ODM with visualization, storage and data analysis functionalities.

The minimum requirements for running the software are a 64bit CPU manufactured on or after 2010, 20 GB of disk space and 4 GB RAM, however much better specifications are needed to process the average amount of images in these projects. In the ODM documentation, the recommended requirements are latest Generation CPU 100 GB of disk space and 16 GB RAM and a chart with minimum RAM requirements per number of images is also referred (See figure).

### Installation

A detailed installation tutorial can be found [here](https://docs.opendronemap.org/installation/). This already a very detailed guide and applicable to many different hardware and software arrangements.

Take into account that before installing ODM you need to install 

* Python [latest version 3](https://www.python.org/downloads/windows/)

* [Git](https://git-scm.com/downloads)

* [Docker Toolboxs](https://github.com/docker/toolbox/releases/download/v18.09.3/DockerToolbox-18.09.3.exe) or [Docker for Windows](https://download.docker.com/win/stable/Docker%20for%20Windows%20Installer.exe) if you have Windows 10 Professional or a newer version 


### Creating the point cloud with ODM

To open WebODM, first launch Docker Desktop and look for the container called webapp as shown in the figure. 

<figure style="text-align:center;">
    <img src="./Images/Screenshot 2023-04-22 145312.png" alt="WebODM" />
    <figcaption style="font-style:italic;">
         WebODM Screenshot, source: own elaboration
    </figcaption>
</figure>



By clicking in the port `8000:8000` the web application will be launched. Once in the dashboard, click `+ add project` botton and insert the details of the project. Select images and in the options sections chose default which will only produce a point cloud (Digital Surface Model) with the selected images.

Once the images are processed the results will be available to explore like in the following figure, 

<figure style="text-align:center;">
    <img src="./Images/Screenshot 2023-04-22 152052.png" alt="WebODM" />
    <figcaption style="font-style:italic;">
         WebODM results, source: own elaboration
    </figcaption>
</figure>


To download the point cloud click on the task and then on `Download Assets` and select point cloud, some of the most common file extensions are LAS or LAZ (compressed version of LAS) and recognized by CloudCompare and Potree. 

<figure style="text-align:center;">
    <img src="./Images/Screenshot 2023-04-22 151622.png" alt="WebODM" />
    <figcaption style="font-style:italic;">
         WebODM downloading the point cloud, source: own elaboration
    </figcaption>
</figure>


Another posibility to process the images is the use of the software [Viswebdrone](https://github.com/Nathalieguimaraes/VisWebDrone) (visual web platform for drone-based imagery analysis) is a web application for photogrammetric processing based on open-source technologies., which uses MicMac (an open source software for photogrammetry developed by the French National Geographic Institute) for processing the point clouds, it does not have however, many resources for non developers. More information [here](https://www.mdpi.com/2220-9964/9/11/679).



Mention Viswebdrone as an option to explore in the future.
